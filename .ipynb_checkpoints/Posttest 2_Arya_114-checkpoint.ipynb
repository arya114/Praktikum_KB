```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

Menampilkan jumlah record, nama attribute, jumlah attribute dengan tipe objek, jumlah attribute dengan tipe
angka, jumlah attribute yang memiliki nilai null.

```python
df.info()
jumlah = len(df)
attribut = df.columns
objek = len(df.select_dtypes(include=['object']).columns)
float = len(df.select_dtypes(include=['float64']).columns)
angka = len(df.select_dtypes(include=['int64']).columns)
null  = len(df.columns[df.isna().sum() > 0])

dictionary_data = {
    "Jumlah Record      ": jumlah,
    "Nama Attribut   ": attribut,
    "Jumlah Attribut ": objek,
    "Jumlah Float    ": float,
    "Jumlah Angka    ": angka,
    "Jumlah Null     ": null,
}

pd.Series(dictionary_data)
```
